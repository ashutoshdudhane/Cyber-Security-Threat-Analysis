```python
import pandas as pd
```

```python
#Load dataset
df = pd.read_csv("updated_cybersecurity_attacks.csv")
df.head()
```

```python
# My dataset from kaggle was already cleaned
df.fillna(0,inplace=True)
```

```python
#Analyze threats per system
df['Device/OS'].value_counts()
```

```python
#Calculate moving average of attacks
daily_attacks = df.groupby(df['Timestamp'].dt.date).size()
moving_avg = daily_attacks.rolling(window=7).mean()

moving_avg.dropna()
```

```python
#Segment threats by severity
df[['Attack Type','Severity Level']].value_counts()
```

```python
#Analyze attack patterns
df['Attack Type'].value_counts()
```

```python
#Identify High risk systems
df[['Device/OS']].value_counts().sort_values(ascending=False)
```
